```python
class Chocolate:
  def __init__ (self, price:float, additive:str=None):

    if price <= 0:
      raise ValueError("Цена батончика должна быть больше нуля")

    self.price=price
    self.additive=additive

  def show_my_chocolate (self):
    if self.additive != None:
      print('Шоколадный батончик, добавка:', self.additive)
    else:
      print('Обычный шоколадный батончик')

  def __str__(self):
    return f"Шоколадный батончик, {self.price} рублей"



chocolate1=Chocolate(12,'Изюм') #Пример 1
print(chocolate1)
chocolate1.show_my_chocolate()
print()

chocolate2=Chocolate(12) #Пример 2
print(chocolate2)
chocolate2.show_my_chocolate()
print()

chocolate3=Chocolate(0,'Орехи') #Пример 3
print(chocolate3)
chocolate3.show_my_chocolate()
print()
```
